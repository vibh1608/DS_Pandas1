```python
#importing the libraries
import numpy as np
import pandas as pd

#How-to-count-distance-to-the-previous-zero

lst=[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]

#creating dataFrame of values in lst
df = pd.DataFrame({'X':lst})

#creating new column 'Y' in df
df['Y']=0

#initialising distance value to be -1
distance=-1

#running loop for each row in df
for i in range(0,len(lst)):
    
    #checking if value in column 'X' is zero then set distance to zero
    if df.loc[i]['X'] == 0:
        distance = 0
        
    #for initial case writing the index of value if no zero found yet
    elif distance == -1:
        df.loc[i]['Y'] = i+1
        
    #else increasing distance value and copying into the column 'Y'
    else:
        distance = distance+1
        df.loc[i]['Y'] = distance

#printing the dataframe
df
```

```python
#Creating a DatetimeIndex that contains each business day of 2015  
dti = pd.date_range(start='2015-01-01', end='2015-12-31', freq='B') 

#using it to index a Series of random numbers
s = pd.Series(np.random.rand(len(dti)), index=dti)

#sum of the values in s for every Wednesday
o1 = s[dti.weekday == 2].sum() 
print("The sum of the values in s for every Wednesday is {}".format(o1),end='\n\n')

#Average For each calendar month
o2 = s.resample('M').mean()
print("Average For each calendar month is: ")
print(o2,end='\n\n')

#The date on which the highest value occurred
o3 = s.groupby(pd.Grouper(freq='4M')).idxmax()
print("The date on which the highest value occurred :")
print(o3)
```
